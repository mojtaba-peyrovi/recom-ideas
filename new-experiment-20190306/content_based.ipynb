```python
vals_to_replace = {'YES':'1', 'NO':'0'}
item_feature['facility_gym'] =item_feature['facility_gym'].map(vals_to_replace)
item_feature['facility_gym'] = item_feature['facility_gym'].fillna(0)
        



item_feature['facility_swimming'] =item_feature['facility_swimming'].map(vals_to_replace)
item_feature['facility_swimming'] = item_feature['facility_swimming'].fillna(0)

item_feature['airport_shuttle'] =item_feature['airport_shuttle'].map(vals_to_replace)
item_feature['airport_shuttle'] = item_feature['airport_shuttle'].fillna(0)

item_feature['facility_wifi'] =item_feature['facility_wifi'].map(vals_to_replace)
item_feature['facility_wifi'] = item_feature['facility_wifi'].fillna(0)

item_feature['facility_parking'] =item_feature['facility_parking'].map(vals_to_replace)
item_feature['facility_parking'] = item_feature['facility_parking'].fillna(0)

rating_mean = item_feature['trust_rating_value'].mean()

item_feature['trust_rating_value'] = item_feature['trust_rating_value'].fillna(rating_mean)

item_feature['d37_hotel_category_name'] = item_feature['d37_hotel_category_name'].str.replace('3-star', '3')
item_feature['d37_hotel_category_name'] = item_feature['d37_hotel_category_name'].str.replace('4-star', '4')
item_feature['d37_hotel_category_name'] = item_feature['d37_hotel_category_name'].str.replace('5-star', '5')
item_feature['d37_hotel_category_name'] = item_feature['d37_hotel_category_name'].str.replace('2-star', '2')
item_feature['d37_hotel_category_name'] = item_feature['d37_hotel_category_name'].str.replace('1-star', '1')

item_feature = item_feature.drop('d37_hotel_category_name', axis=1)
```
